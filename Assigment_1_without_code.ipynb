## Exercise 4
"I am interested in the house with 3 bedrooms, 2 bathrooms and 2 floors"  
a) Compute the basic statistics of apartments fulfilling above constraints.  
b) Distinct accepted apartments on any of previously created plots.  
c) Find such apartments with best price to footage ratio.  

```python
#a)
#Your code:

# 2 floors - podejrzewam, Å¼e chodzi o to, Å¼eby sqrt_basement =/= 0
# name_df.bathrooms.to_numpy() # =2 
# name_df.bedrooms.to_numpy() # =3
# name_df.sqft_basement.to_numpy() # =/=0
# z caÅ‚ego dataframe wybraÄ‡ tylko elementy speÅ‚niajÄ…ce powyÅ¼sze warunki
# NakÅ‚adam na dataframe kolejne warunki:
war1_df = name_df.loc[name_df['sqft_basement'] > 0]
war2_df = war1_df.loc[war1_df['bathrooms'] == 2]
# Ostatecznie otrzymujÄ™ dataframe speÅ‚niajÄ…cy wszystkie warunki
war_df = war2_df.loc[war2_df['bedrooms'] == 3]
war_df.describe()

#Hint: pandas.DataFrame.loc
```

```python
#b)
#Your code: 

name_df['filtered'] = np.where((name_df['sqft_basement'] > 0) & (name_df['bathrooms'] == 2) & (name_df['bedrooms'] == 3), 30, 4)
#name_df.head()
a = name_df['grade']
b = name_df['price']
plt.set_cmap('flag')
#sizes = [3 if name_df['filtered']==1 else 7]
plt.scatter(a,b,c=name_df["filtered"], s=name_df['filtered']) 
```

```python
#c)
#Your code:
# best price to footage(="sqft_living") ratio
name_df["ratio"] = name_df['price']/name_df['sqft_living']
best_price = name_df.loc[name_df['ratio'] < 182]
#Hint: .loc, .sort_values(), .index
```
